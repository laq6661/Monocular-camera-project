```python
import sys
import numpy as np
import os
datafile = r"/prgramming data/human_sk_beida/PKU_Skeleton_Renew/"
labelfile= r"/prgramming data/human_sk_beida/Train_Label_PKU_final/"
filename = os.listdir(labelfile)

def new_label_maker(labelfilename) :
    label = []
    new_label=[]  
    with open(labelfilename,'r') as fi:
        for line2 in fi:
            label.append(list(line2.strip('\n').strip().split(',')))
    for i in range(len(label)):
        ml = []
        for j in range(len(label[0])):
            ml.append(int(label[i][j]))
        new_label.append(ml)
    return new_label

def real_label_maker(new_label):
    real_label = []
    for i in new_label:
        num = i[2]-i[1]
        for j in range(0,num,1):
            real_label.append(i[0])
    #print("real_label shape:",len(real_label))
    return real_label

def real_data_maker(datafilename,new_label):
    data=[]
    new_data=[]
    real_data=[]
    with open(datafilename,'r') as f:
        for line in f:
            data.append(list(line.strip('\n').strip().split(' ')))
  
    for i in range(len(data)):
        md = []
        for j in range(len(data[0])):
            md.append(float(data[i][j]))
        new_data.append(md)
    #print("new_data shape:",len(new_data),len(new_data[0]))
    for i in new_label:
        for j in range(i[1],i[2],1):
            real_data.append(new_data[j])
    #print("real_data shape:",len(real_data),len(real_data[0]))
    return real_data
```

```python
label = []
data = []
count = 0
for name in filename:
    print("----------------now is deal with  ",name," file")
    labelfilename = labelfile+name
    datafilename = datafile+name
    new_label = new_label_maker(labelfilename)
    label = label+(real_label_maker(new_label))
    data = data +(real_data_maker(datafilename,new_label))
#     count +=1
#     if(count >10):
#         break
data = np.array(data)
label = np.array(label)
data = data.reshape(-1,50,3)
print(np.shape(data))
print(np.shape(label))
print(data[0])
print(label[0])
```

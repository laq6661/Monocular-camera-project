```python
import sys
import numpy as np
import os
from scipy import io
```

```python
def onefile(mat_t):
  m_data = []
  num = len(mat_t)-len(mat_t)%3
  #num = len(mat_t)
  for i in range(0,num,1):
    onetime=[]
    for j in range(0,20,1):
      onetime.append([mat_t[i,0,j],mat_t[i,1,j],mat_t[i,2,j]])
    m_data.append(onetime)

  # m_data = np.array(m_data)   
  # print(m_data.shape)
  # print(m_data[0:2,:,:])
  return m_data

def label_maker(name,mat_t):
  new_label = []
  num = len(mat_t)-len(mat_t)%3
  num=num/3
  #num = len(mat_t)
  for i in range(0,int(num),1):
        new_label.append(name.split('_')[0])
  return new_label
```

```python
data = [] 
label = []
datafilename = r'H:/LI project/3d sk detect/UTD-MHAD/UTD-MHAD_skeleton/a1_s2_t1_skeleton.mat'
mat = io.loadmat(datafilename)
mat_t = np.transpose(mat['d_skel'])
data = data + (onefile(mat_t))
label = label+label_maker('a1_s2_t1_skeleton.mat',mat_t)
data = np.array(data)
label = np.array(label)
data = data.reshape(-1,3,20,3)
print(data.shape)
print(data[0:2])
print(label.shape)
print(label[0:2])
```

```python
trainX = []
trainY =[]
testX = []
testY=[]
dataFile = r'H:/LI project/3d sk detect/UTD-MHAD/UTD-MHAD_skeleton/'
filename = os.listdir(dataFile)
for name in filename:
    print("----------------now is deal with  ",name," file")
    x = name.split("_")[1]
    if(x =="s1"or x=='s3'or x=='s5' or x=='s7'):
        datafilename = dataFile+name
        mat = io.loadmat(datafilename)
        mat_t = np.transpose(mat['d_skel'])
        trainX = trainX + (onefile(mat_t))
        trainY = trainY+label_maker(name,mat_t)
    else:
        datafilename = dataFile+name
        mat = io.loadmat(datafilename)
        mat_t = np.transpose(mat['d_skel'])
        testX = testX + (onefile(mat_t))
        testY = testY+label_maker(name,mat_t)
trainX = np.array(trainX)
trainY = np.array(trainY)
testX = np.array(testX)
testY = np.array(testY)

trainX = trainX.reshape(-1,3,20,3)
testX = testX.reshape(-1,3,20,3)
print('trainx',trainX.shape)
print('trainy',trainY.shape)
print('testx',testX.shape)
print('testy',testY.shape)
```
